## Generating the ResNet

```python
from tensorflow.keras.layers import *

from tensorflow.keras.initializers import glorot_uniform
def resBlock(X, filters, stage):
    ########################
    # Convolution Block    #
    ########################
    X_copy = X
    filter1, filter2, filter3 = filters
    
    # Main Path
    X = Conv2D(filters = filter1, kernel_size = (1,1), strides = (1,1), name ='res_'+str(stage)+'_conv_a', 
               kernel_initializer = glorot_uniform(seed = 0))(X)
    X = MaxPool2D(pool_size = (2,2))(X)
    X = BatchNormalization(axis = 3, name='bn_'+str(stage)+'_conv_a')(X)
    X = Activation('relu')(X) 
    
    X = Conv2D(filters = filter2, kernel_size = (3, 3), strides = (1,1), padding = 'same',
               name ='res_'+str(stage)+'_conv_b', kernel_initializer= glorot_uniform(seed = 0))(X)
    X = BatchNormalization(axis = 3, name='bn_'+str(stage)+'_conv_b')(X)
    X = Activation('relu')(X) 
    
    X = Conv2D(filters = filter3, kernel_size = (1,1), strides = (1,1), name ='res_'+str(stage)+'_conv_c',
               kernel_initializer= glorot_uniform(seed = 0))(X)
    X = BatchNormalization(axis = 3, name='bn_'+str(stage)+'_conv_c')(X)
    
    # Short Path
    X_copy = Conv2D(filters = filter3, kernel_size = (1,1), strides = (1,1), name ='res_'+str(stage)+'_conv_short',
               kernel_initializer= glorot_uniform(seed = 0))(X_copy)
    X_copy = MaxPool2D(pool_size = (2,2))(X_copy)
    X_copy = BatchNormalization(axis = 3, name='bn_'+str(stage)+'_conv_short')(X_copy)
    
    #Addtion and ReLU
    X = Add()([X, X_copy])
    X = Activation('relu')(X)

    
    
    ########################
    # Identity Block 1     #
    ########################
    X_copy = X
    
    # Main Path
    X = Conv2D(filters = filter1, kernel_size = (1,1), strides = (1,1), name ='res_'+str(stage)+'_identity_1_a',
               kernel_initializer= glorot_uniform(seed = 0))(X)
    X = BatchNormalization(axis = 3, name='bn_'+str(stage)+'_identity_1_a')(X)
    X = Activation('relu')(X) 
    
    X = Conv2D(filters = filter2, kernel_size = (3, 3), strides = (1,1), padding = 'same',
               name ='res_'+str(stage)+'_identity_1_b', kernel_initializer= glorot_uniform(seed = 0))(X)
    X = BatchNormalization(axis = 3, name='bn_'+str(stage)+'_identity_1_b')(X)
    X = Activation('relu')(X) 
    
    X = Conv2D(filters = filter3, kernel_size = (1,1), strides = (1,1), name ='res_'+str(stage)+'_identity_1_c',
               kernel_initializer= glorot_uniform(seed = 0))(X)
    X = BatchNormalization(axis = 3, name='bn_'+str(stage)+'_identity_1_c')(X)
    
    #Addtion and ReLU
    X = Add()([X, X_copy])
    X = Activation('relu')(X)
    
    
    ########################
    # Identity Block 2     #
    ########################
    X_copy = X
    
    # Main Path
    X = Conv2D(filters = filter1, kernel_size = (1,1), strides = (1,1), name ='res_'+str(stage)+'_identity_2_a',
               kernel_initializer= glorot_uniform(seed = 0))(X)
    X = BatchNormalization(axis = 3, name='bn_'+str(stage)+'_identity_2_a')(X)
    X = Activation('relu')(X) 
    
    X = Conv2D(filters = filter2, kernel_size = (3, 3), strides = (1,1), padding = 'same',
               name ='res_'+str(stage)+'_identity_2_b', kernel_initializer= glorot_uniform(seed = 0))(X)
    X = BatchNormalization(axis = 3, name='bn_'+str(stage)+'_identity_2_b')(X)
    X = Activation('relu')(X) 
    
    X = Conv2D(filters = filter3, kernel_size = (1,1), strides = (1,1), name ='res_'+str(stage)+'_identity_2_c',
               kernel_initializer= glorot_uniform(seed = 0))(X)
    X = BatchNormalization(axis = 3, name='bn_'+str(stage)+'_identity_2_c')(X)
    
    #Addtion and ReLU
    X = Add()([X, X_copy])
    X = Activation('relu')(X)
    
    return X
     
    
```

```python
from tensorflow.keras.models import Model

inputShape = (96, 96, 1)

# Creating input tensor with `inputShape` shape
x_input = Input(inputShape)

# Zero-Padding
x = ZeroPadding2D((3,3))(x_input)

# Stage 1
x = Conv2D(filters =64, kernel_size = (7,7), strides = (2,2), name="Convolution1", 
           kernel_initializer= glorot_uniform(seed = 0))(x)
X = BatchNormalization(axis =3, name="BatchNorm1")(x)
x = Activation("relu")(x)
x = MaxPooling2D((3,3), strides = (2,2))(x)


# Stage 2
x = resBlock(x, filters=(64, 64, 256), stage=2)

# Stage 3
x = resBlock(x, filters=(128, 128, 512), stage=3)

# Average Pooling
x = AveragePooling2D(pool_size = (2,2), name="AvgPooling")(x)

# Final Layer
x = Flatten()(x)
x = Dense(units = 4096, activation='relu')(x)
x = Dropout(0.2)(x)
x = Dense(units = 2048, activation='relu')(x)
x = Dropout(0.1)(x)
x = Dense(units=30, activation = 'relu')(x)


keyFacialPoints_model = Model(inputs = x_input, outputs = x)
keyFacialPoints_model.summary()
```

```python
from tensorflow.keras import optimizers

adam = optimizers.Adam(learning_rate = 0.0001, beta_1 = 0.9, beta_2 = 0.999, amsgrad = False)
keyFacialPoints_model.compile(loss = "mean_squared_error", optimizer=adam, metrics = ['accuracy'])
```

```python
from tensorflow.keras.callbacks import ModelCheckpoint

checkpointCB = ModelCheckpoint(filepath="Dataset/keyFacialPoints_weights.hdf5", verbose = 1, save_best_only = True)
history = keyFacialPoints_model.fit(X_train, y_train, batch_size = 64, epochs = 250, validation_split = 0.1, callbacks=[checkpointCB])
```

# Part 2: Facial Expression Detection

```python
facialExpression_df = pd.read_csv("Dataset/icml_face_data.csv")
```

```python
facialExpression_df
```

```python
# len(facialExpression_df['pixels'][0].split())  # 2304 - > 48x48
facialExpression_df.pixels = facialExpression_df.pixels.apply(lambda x: np.fromstring(x, dtype = np.float32, sep = ' ').reshape(48,48))
```

```python
facialExpression_df.pixels[0].shape
```

```python
# Resizing 48x48 images to match 96x96 resolution used in Key Facialpoints detection
import cv2
facialExpression_df.pixels = facialExpression_df.pixels.apply(lambda x: cv2.resize(x ,dsize = (96, 96), interpolation = cv2.INTER_CUBIC))
```

```python
facialExpression_df.shape
```

```python
facialExpression_df.isnull().sum()
```

```python
label_to_text = {0:'anger', 1:'disgust', 2:'sad', 3:'happiness', 4: 'surprise'}
```

```python
label_to_text.values()
```

```python
emotions = [0, 1, 2, 3, 4]

for i in emotions:
  data = facialExpression_df[facialExpression_df['emotion'] == i][:1]
  img = data['pixels'].item()
  img = img.reshape(96, 96)
  plt.figure()
  plt.title(label_to_text[i])
  plt.imshow(img, cmap = 'gray')
```

```python
facialExpression_df.emotion.value_counts()
```

```python
plt.figure(figsize = (15,15))
sns.barplot(x = facialExpression_df.emotion.value_counts().index, y =facialExpression_df.emotion.value_counts())
```

```python
from keras.utils import to_categorical

X = np.stack(facialExpression_df.pixels, axis = 0)
X /= 255
X = X.reshape(24568, 96, 96, 1)

y = to_categorical(facialExpression_df.emotion)
```

```python
X.shape, y.shape
```

```python
#Creating Validation data as well, because we will be feeding image generator data to the model

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.25)
X_test, X_val, y_test, y_val = train_test_split(X_test, y_test, test_size = 0.5)
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

imageGenerator = ImageDataGenerator(rotation_range=15, width_shift_range=0.1, height_shift_range=0.1,
    shear_range=0.1, zoom_range=1.0, horizontal_flip = True)
```

```python
input_shape = (96, 96, 1)

# Input tensor shape
x_input = Input(input_shape)

# Zero-padding
x = ZeroPadding2D((3, 3))(x_input)

# 1 - stage
x = Conv2D(64, (7, 7), strides= (2, 2), name = 'conv1', kernel_initializer= glorot_uniform(seed = 0))(x)
x = BatchNormalization(axis =3, name = 'bn_conv1')(x)
x = Activation('relu')(x)
x = MaxPooling2D((3, 3), strides= (2, 2))(x)

# 2 - stage
x = resBlock(x, filters = [64, 64, 128], stage= 2)

# 3 - stage
x = resBlock(x, filters = [128, 128, 256], stage= 3)

# 4 - stage
x = resBlock(x, filters = [256, 256, 512], stage= 4)

# Average Pooling
x = AveragePooling2D((2, 2), name = 'Average_Pooling')(x)

# Final layer
x = Flatten()(x)
x= Dense(5, activation = 'softmax', name = 'Dense_final', kernel_initializer= glorot_uniform(seed=0))(x)

facialExpression_model = Model( inputs= x_input, outputs = x, name = 'FacialExpressionDetection')

facialExpression_model.summary()
```

```python
facialExpression_model.compile(optimizer='Adam', loss = 'categorical_crossentropy',metrics = ['accuracy'])
```

```python
#Defining Call backs
from tensorflow.keras.callbacks import EarlyStopping

#Early stopping to stop training if performance is not improving
earlyStoppingCB = EarlyStopping(monitor = 'val_loss', patience = 70, mode='min', verbose = 1)
checkpointCB = ModelCheckpoint(filepath="Dataset/facialExpressions_weights.hdf5", verbose = 1, save_best_only = True)

```

```python
history = facialExpression_model.fit(imageGenerator.flow(X_train, y_train, batch_size = 64),
                                validation_data=(X_val, y_val), steps_per_epoch = len(X_train)//64,
                                epochs = 1000, callbacks = [earlyStoppingCB, checkpointCB])
```

```python
# saving the model architecture to json file for future use

facialExpression_model_json = facialExpression_model.to_json()
with open("Dataset/facialExpressions.json","w") as json_file:
  json_file.write(facialExpression_model_json)
```

```python
score = facialExpression_model.evaluate(X_test, y_test)
print("Val Loss: {}, Accuracy: {}".format(score[0], score[1]))
```
